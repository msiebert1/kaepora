```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../data/info_files/hsiao_lc_template.dat', unpack = True)
sn20zmi_file = open('../../Observing/2020zmi_data.snana.txt')
sn20zmi_data = sn20zmi_file.readlines()
times = []
times_r = []
gs = []
rs = []
for line in sn20zmi_data:
    if (line.split()[2] == 'g' or line.split()[2] == 'g-ZTF' or line.split()[2] == 'gp') and line.split()[5] != 'nan':
        print line
        times.append(float(line.split()[1]))
        gs.append(float(line.split()[5]))
    if (line.split()[2] == 'r' or line.split()[2] == 'r-ZTF' or line.split()[2] == 'rp') and line.split()[5] != 'nan':
        print line
        times_r.append(float(line.split()[1]))
        rs.append(float(line.split()[5]))
```

```python
print times
print gs
print rs
times = np.asarray(times)
times_r = np.asarray(times_r)
gs = np.asarray(gs)
rs = np.asarray(rs)

inds  = np.argsort(times)
times = times[inds]
gs = gs[inds]

inds_r  = np.argsort(times_r)
times_r = times_r[inds_r]
rs = rs[inds_r]

print times
print times_r
print gs
print rs

plt.scatter(times,gs)
plt.gca().invert_yaxis()
plt.show()
plt.scatter(times_r,rs)
plt.gca().invert_yaxis()
plt.show()
```
