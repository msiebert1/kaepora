```python
composites = query_db.main(len(query_strings), query_strings)
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
%matplotlib notebook
# fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
fig, ax = plt.subplots(1,1)
# fig = plt.figure(num = 1, dpi = 100, figsize = [20,10], facecolor = 'w')
plt.xlim([3000,10000])
plt.ylim([0.,12.e-15])
s_m = query_db.make_colorbar(composites)
cb = plt.colorbar(s_m, ax = fig.axes)
phases = []
for comp in composites:
    phases.append(np.average(comp.phase_array[comp.x1:comp.x2]))
    
for f in range(len(composites)):
    ax.plot(composites[f].wavelength[composites[f].x1:composites[f].x2], 
            composites[f].flux[composites[f].x1:composites[f].x2], color = s_m.to_rgba(phases[f]))
    fig.canvas.draw()
    time.sleep(.2)
```
